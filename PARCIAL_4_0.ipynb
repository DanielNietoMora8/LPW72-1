**1. Herencia básica: Figuras Geométricas**

Implementa un programa que modele un sistema de figuras geométricas:

+ Crea una clase base llamada Figura con:

  • Un método calcular area() que retorne 0 (debe ser sobreescrito por las clases hijas).

+ Crea dos clases hijas:

   • Rectangulo: Calcula el área de un rectáangulo (base * altura).

   • Triangulo: Calcula el área de un triángulo (0.5 * base * altura).

+ Implementa una función principal que cree un Rectangulo y un Triangulo, solicite
al usuario los valores necesarios, y muestre el ´area de cada figura.

```python
# Clase base Figura
class Figura:
    def calcular_area(self):
        return 0


# Clase_hija_1 Rectangulo, herencia de Figura
class Rectangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def calcular_area(self):
        # Área del rectángulo = base * altura
        return self.base * self.altura


# Clase_hija_2 Triandulo, herencia de Figura
class Triangulo(Figura):
    def __init__(self, base, altura):
        self.base = base
        self.altura = altura

    def calcular_area(self):
        # Área del triángulo = 0.5 * base * altura
        return 0.5 * self.base * self.altura


# Función principal e interacción con usuario
def main():
    # Solicitar datos al usuario para el rectángulo
    print("Por favor ingrese datos para el rectángulo:")
    base_rectangulo = float(input("Base rectángulo: "))
    altura_rectangulo = float(input("Altura rectángulo: "))

    # Crear una instancia de Rectangulo
    rectangulo = Rectangulo(base_rectangulo, altura_rectangulo)

    # Solicitar datos al usuario para el triángulo
    print("\nPor favor ingrese datos para el triángulo:")
    base_triangulo = float(input("Base triángulo: "))
    altura_triangulo = float(input("Altura triángulo: "))

    # Crear una instancia de Triangulo
    triangulo = Triangulo(base_triangulo, altura_triangulo)

    # Mostrar los resultados
    print("\nÁrea del rectángulo:", rectangulo.calcular_area())
    print("Área del triángulo:", triangulo.calcular_area())


# Ejecutar función principal
if __name__ == "__main__":
    main()
```

**2.** **Grafica de audio con librosa y matplotlib (30 %)**

Desarrolla un programa siguiendo las siguientes indicaciones:

+ Cargue un audio con una duración de entre 10 y 20 segundos en un formato legible por librosa.

+ Cree una clase con la ruta del audio como base, es decir, la ruta es el atributo principal y con el se harán los siguientes pasos.

+ Use librosa para cargar el archivo y cree un método en la clase para extraer y
retornar la forma de onda (waveform) en un array.

+ Cree un método que genere una gráfica de la forma de onda usando matplotlib,
plotly o seaborn.

+ Cree un método que guarde la gráfica en formato .png en la misma carpeta del
script.

+ Implemente un método que utilice alguna otra función de librosa de su elección

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np
import os

# Clase para tratamiento del audio
class AudioProcessor:
    def __init__(self, audio_path):
        self.audio_path = audio_path
        self.audio_data, self.sr = librosa.load(audio_path) # Cargar audio con librosa

    def get_waveform(self):
        """
        Retornar la forma de onda del audio.
        """
        return self.audio_data

    def plot_waveform(self):
        """
        Genera una gráfica de la forma de onda usando matplotlib.
        """
        plt.figure(figsize=(10, 3))
        librosa.display.waveshow(self.audio_data, sr=self.sr)
        plt.title("Forma de onda del audio")
        plt.xlabel("Tiempo (s)")
        plt.ylabel("Amplitud")
        plt.grid(True)

    def save_waveform_plot(self):
        """
        Guardar la gráfica de la forma de onda como un archivo PNG.
        """
        plot_filename = os.path.join(os.path.dirname(self.audio_path), 'waveform_plot.png')
        plt.figure(figsize=(10, 3))
        librosa.display.waveshow(self.audio_data, sr=self.sr)
        plt.title("Forma de onda del audio")
        plt.xlabel("Tiempo (s)")
        plt.ylabel("Amplitud")
        plt.grid(True)
        plt.savefig(plot_filename)
        print(f"Gráfica guardada como: {plot_filename}")

    def additional_analysis(self):
        """
        Análisis adicional usando librosa. Espectrograma.
        """
        # Calculando el espectrograma de amplitud
        D = librosa.amplitude_to_db(librosa.stft(self.audio_data), ref=np.max)

        plt.figure(figsize=(10, 3))
        librosa.display.specshow(D, sr=self.sr, x_axis='time', y_axis='log')
        plt.colorbar(format='%+2.0f dB')
        plt.title('Espectrograma')
        plt.xlabel('Tiempo (s)')
        plt.ylabel('Frecuencia (Hz)')
        plt.show()


# Función principal
def main():
    # Definir la ruta al archivo de audio (modificar según corresponda)
    audio_path = '/content/bird-sounds-241394.mp3'

    # Crear una instancia de AudioProcessor
    audio_processor = AudioProcessor(audio_path)

    # Obtener la forma de onda
    waveform = audio_processor.get_waveform()
    print("Forma de onda cargada, longitud del audio:", len(waveform), "samples.")

    # Graficar la forma de onda
    audio_processor.plot_waveform()

    # Guardar la gráfica de la forma de onda como un archivo PNG
    audio_processor.save_waveform_plot()

    # Análisis adicional (espectrograma)
    audio_processor.additional_analysis()


if __name__ == "__main__":
    main()
```

**3. Procesamiento de im´agenes con OpenCV (25 %)**

+ Crea un programa que permita al usuario aplicar un filtro de detección de bordes a una imagen -¿No se requiere utilizar POO:

  • Cargue una imagen (.png o .jpg).

  • Usar OpenCV para convertirla a escala de grises y luego aplicar el filtro de bordes de Canny (esto lo hace OpenCV, hay que leer la documentación).

  • Mostrar la imagen resultante y guardarla en un nuevo archivo.

```python
import cv2
import matplotlib.pyplot as plt

# Cargar imagen
image_path = input("/content/Captura_pai.PNG")

# Leer imagen usando OpenCV
image = cv2.imread(image_path)

# Convertir imagen a escala de grises
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Filtro de bordes de Canny
edges = cv2.Canny(gray_image, threshold1=100, threshold2=200)

# Mostrar ambas imagenes
plt.figure(figsize=(10, 5))

# Imagen original
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Imagen Original')
plt.axis('off')  # No mostrar los ejes

# Imagen con bordes detectados
plt.subplot(1, 2, 2)
plt.imshow(edges, cmap='gray')
plt.title('Imagen con bordes detectados (Filtro Canny)')
plt.axis('off')  # No mostrar los ejes

# Mostrar las imágenes
plt.show()

# Guardar la imagen con los bordes detectados
output_path = "bordes_detectados.png"
cv2.imwrite(output_path, edges)
print(f"La nueva imagen con bordes detectados se ha guardado satisfactoriamente como: {output_path}")
```
